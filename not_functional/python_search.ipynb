```python
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
import numpy as np
```

```python
# Extract data from amazon.csv
amazon_metadata_df = pd.read_csv('resources/csv/amazon_prime_titles.csv')
amazon_metadata_df.head()
```

```python
amazon_df = amazon_metadata_df[["title", "country","release_year","description"]]
amazon_df.head(10)
```

```python
# Extract data from disney.csv
disney_metadata_df = pd.read_csv('resources/csv/disney_plus_titles.csv')
disney_metadata_df.head()
```

```python
disney_df = disney_metadata_df[["title", "country","release_year","description"]]
disney_df.head(10)
```

```python
# Extract data from hulu.csv
hulu_metadata_df = pd.read_csv('resources/csv/hulu_titles.csv')
hulu_metadata_df.head()
```

```python
hulu_df = hulu_metadata_df[["title", "country","release_year","description"]]
hulu_df.head(10)
```

```python
# Extract data from netflix.csv
netflix_metadata_df = pd.read_csv('resources/csv/netflix_titles.csv')
netflix_metadata_df.head()
```

```python
netflix_df = netflix_metadata_df[["title", "country","release_year","description"]]
netflix_df.head(10)
```

```python
platform = input("Enter movie or tv show to know which streaming platform has it")
print("Username is: " + platform)
```

```python
amazon_title = amazon_df.loc[amazon_df['title'].str.contains( platform , case=False)]
amazon_title
```

```python
disney_title = disney_df.loc[disney_df['title'].str.contains( platform , case=False)]
disney_title
```

```python
hulu_title = hulu_df.loc[hulu_df['title'].str.contains( platform , case=False)]
hulu_title
```

```python
netflix_title = netflix_df.loc[netflix_df['title'].str.contains( platform , case=False)]
netflix_title
```

```python
count_amazon= amazon_title["title"].count()
count_disney= disney_title["title"].count()
count_hulu= hulu_title["title"].count()
count_netflix= netflix_title["title"].count()
```

```python
if count_amazon > 0:
    print(platform + " is avaliable on Amazon Prime")
else:
    print(platform + " is NOT avaliable on Amazon Prime")
```

```python
if count_disney > 0:
    print(platform + " is avaliable on Disney Plus")
else:
    print(platform + " is NOT avaliable on Disney Plus")
```

```python
if count_hulu > 0:
    print(platform + " is avaliable on Hulu")
else:
    print(platform + " is NOT avaliable on Hulu")
```

```python
if count_netflix > 0:
    print(platform + " is avaliable on Netflix")
else:
    print(platform + " is NOT avaliable on Netflix")
```

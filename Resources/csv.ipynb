```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Reading in csv as dataframes
df_ntfx = pd.read_csv('Resources/netflix_titles.csv')
df_amzn = pd.read_csv('Resources/amazon_prime_titles.csv')
df_dsny = pd.read_csv('Resources/disney_plus_titles.csv')
df_hulu = pd.read_csv('Resources/hulu_titles.csv')

df_ntfx.head()
#df_amzn.head()
#df_dsny.head()
#df_hulu.head()
```

```python
# Filtering dataframes only by relevant columns
df_ntfx = df_ntfx[['type','title','date_added','release_year','rating','duration','listed_in']]
df_amzn = df_amzn[['type','title','date_added','release_year','rating','duration','listed_in']]
df_dsny = df_dsny[['type','title','date_added','release_year','rating','duration','listed_in']]
df_hulu = df_hulu[['type','title','date_added','release_year','rating','duration','listed_in']]

df_ntfx.head()
#df_amzn.head()
#df_dsny.head()
#df_hulu.head()
```

```python
# Adding new column 'service' and populating with values
df_ntfx['service'] = 'Netflix'
df_amzn['service'] = 'Amazon Prime'
df_dsny['service'] = 'Disney+'
df_hulu['service'] = 'Hulu'

df_ntfx.head()
#df_amzn.head()
#df_dsny.head()
#df_hulu.head()
```

```python
# Creating engine to sqlite
engine = create_engine('sqlite:///streaming_db.sqlite', echo=False)
```

```python
# Converting dataframes to sqlite
df_ntfx.to_sql(name='netflix', con=engine, if_exists='append', index=True)
df_amzn.to_sql(name='amazonprime', con=engine, if_exists='append', index=True)
df_dsny.to_sql(name='disneyplus', con=engine, if_exists='append', index=True)
df_hulu.to_sql(name='hulu', con=engine, if_exists='append', index=True)
```
